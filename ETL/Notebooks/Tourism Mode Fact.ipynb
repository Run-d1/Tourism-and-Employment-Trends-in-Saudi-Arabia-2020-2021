```python
import pandas as pd

folder_path = 'DSS_DATASET_UPDATED.xlsx'
df = pd.read_excel(folder_path, sheet_name = 'O5')
# O2 = Purpose of Trips 
# O3 =  Number by Destination - Region
# O5 = Mode of Departure

df
```

```python
df = df.iloc[3:].reset_index(drop=True)

df
```

```python
df.rename(columns={'Unnamed: 1': 'Year'}, inplace=True)
df.rename(columns={'Unnamed: 2': 'Month'}, inplace=True)

df
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
index = 0
for month in df['Month']:
    if index >= 2:
        if month not in months:
            df = df.drop(index)
    index += 1

df = df.reset_index(drop=True)
df
```

```python
years = []
count = 0
for Num in df['Year']:
    if not pd.isnull(Num):
        years.append(int(Num))
    else:
        if count > 0:
            years.append(int(years[count-1]))
        else:
            years.append(0)
    count += 1

df['Year'] = years

df
```

```python
df['Year'] = df['Year'].astype(str)
df['Month'] = df['Month'].astype(str)

df
```

```python
df['Year - Month'] = df['Year'] + ' - ' + df['Month']

df
```

```python
col = df['Year - Month']
df.drop(labels=['Year - Month'], axis=1, inplace=True)
df.insert(0, 'Year - Month', col)

df
```

```python
df.drop(['(Last update: 26-06-2023)', 'Year', 'Month'], axis=1, inplace=True)

df
```

```python
df = df.transpose()

df
```

```python
df = df.rename(columns={0: 'Mode_of_Departure'})
df = df.rename(columns={1: 'Citizenship'})

df
```

```python
df['Mode_of_Departure'] = df['Mode_of_Departure'].str.split('\n').str[0]
df['Citizenship'] = df['Citizenship'].str.split('\n').str[0]

df
```

```python
df = df.reset_index(drop=True)

df
```

```python
citizenships = ['Saudi', 'Non-Saudi']
index = 0
for val in df['Citizenship']:
    if index >= 1:
        if val not in citizenships:
            df = df.drop(index)
    index += 1
        
df
```

```python
modes = []
count = 0
for val in df['Mode_of_Departure']:
    if not pd.isnull(val):
        modes.append(val)
    else:
        if count > 0:
            modes.append(modes[count-1])
        else:
            modes.append(None)
    count += 1

df['Mode_of_Departure'] = modes

df = df.reset_index(drop=True)
df
```

```python
df = df.iloc[0:7].reset_index(drop=True)

df
```

```python
df['Mode, Citizenship'] = df['Mode_of_Departure'] + ', ' + df['Citizenship']

df
```

```python
col = df['Mode, Citizenship']
df.drop(labels=['Mode, Citizenship'], axis=1, inplace=True)
df.insert(0, 'Mode, Citizenship', col)

df
```

```python
df.drop(['Mode_of_Departure', 'Citizenship'], axis=1, inplace=True)

df
```

```python
df = df.transpose()
df = df.reset_index(drop=True)
df
```

```python
df.columns = df.iloc[0]
df = df[1:].reset_index(drop=True)

df
```

```python
fact_table = df.rename(columns={'0 - Month, 0 - nan': 'Year - Month'})

fact_table
```

```python
fact_table = pd.melt(fact_table, id_vars=['Year - Month'], var_name="Mode, Citizenship", value_name="Number_of_Tourists")

fact_table
```

```python
fact_table['Year'] = fact_table['Year - Month'].str.split(' - ').str[0]
fact_table['Month'] = fact_table['Year - Month'].str.split(' - ').str[1]

fact_table.drop(columns=['Year - Month'], inplace=True)

fact_table
```

```python
quarters = []
for month in fact_table['Month']:
    if month in months[0:3]:
        quarters.append('Q1')
    elif month in months[3:6]:
        quarters.append('Q2')
    elif month in months[6:9]:
        quarters.append('Q3')
    elif month in months[9:]:
        quarters.append('Q4')
    else:
        quarters.append(None)

fact_table['Quarter'] = quarters

fact_table
```

```python
fact_table['Mode_of_Departure'] = fact_table['Mode, Citizenship'].str.split(', ').str[0]
fact_table['Citizenship'] = fact_table['Mode, Citizenship'].str.split(', ').str[1]

fact_table.drop(columns=['Mode, Citizenship'], inplace=True)

fact_table
```

```python
date_dim = pd.read_csv('Tourism_Date_Dim.csv')

date_dim
```

```python
fact_table['Year'] = fact_table['Year'].astype(int) 

fact_table
```

```python
fact_table = pd.merge(fact_table, date_dim, on=['Year', 'Quarter', 'Month'], how='left')

fact_table
```

```python
citizenship_dim = pd.read_csv('Tourism_Citizenship_Dim.csv')

citizenship_dim
```

```python
fact_table = pd.merge(fact_table, citizenship_dim, on='Citizenship', how='left')

fact_table
```

```python
mode_dim = pd.read_csv('Tourism_Mode_of_Departure_Dim.csv')

mode_dim
```

```python
fact_table = pd.merge(fact_table, mode_dim, on='Mode_of_Departure', how='left')

fact_table
```

```python
fact_table = fact_table[['Date_ID', 'Citizenship_ID', 'Mode_of_Departure_ID', 'Number_of_Tourists']]

fact_table
```

```python
fact_table = fact_table.sort_values(by=['Date_ID', 'Citizenship_ID', 'Mode_of_Departure_ID'])

fact_table
```

```python
fact_table['Number_of_Tourists'] = fact_table['Number_of_Tourists'].apply(lambda x: '{:,.3f}'.format(x) if x != 0 else x)

fact_table
```

```python
fact_table['Number_of_Tourists'] = fact_table['Number_of_Tourists'].apply(lambda x: x.replace('.', '') if x != 0 else x)

fact_table
```

```python
fact_table['Number_of_Tourists'] = fact_table['Number_of_Tourists'].astype(str)

fact_table
```

```python
fact_table['Number_of_Tourists'] = fact_table['Number_of_Tourists'].apply(lambda x: x.lstrip('0') if x != '0' else x)

fact_table
```

```python
fact_table['Number_of_Tourists'] = fact_table['Number_of_Tourists'].apply(lambda x: '{:,.0f}'.format(int(x)) if x != '0' else x)

fact_table
```

```python
fact_table.to_csv('Tourism_Mode_of_Departure_Fact.csv', index=False)
```
