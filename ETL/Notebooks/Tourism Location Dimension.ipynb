```python
import pandas as pd

folder_path = 'DSS_DATASET_UPDATED.xlsx'
df = pd.read_excel(folder_path, sheet_name = 'O3')
# O2 = Purpose of Trips 
# O3 =  Number by Destination - Region
# O5 = Mode of Departure

df
```

```python
df = df.transpose()

df
```

```python
col = [3]

dim_table = df[col]

dim_table
```

```python
dim_table.columns = ['Region']

dim_table
```

```python
dim_table = dim_table.dropna().reset_index(drop=True)

dim_table
```

```python
dim_table = dim_table.drop(0)
dim_table = dim_table.drop(dim_table.index[7:])

dim_table
```

```python
dim_table['Region'] = dim_table['Region'].str.split('\n').str[0]

dim_table
```

```python
dim_table = dim_table.reset_index(drop=True)

dim_table
```

```python
def format_region(region):
    words = region.split()
    formatted_words = []
    for word in words:
        if len(word) <= 3:
            formatted_word = word
        else:
            formatted_word = word.capitalize()
        formatted_words.append(formatted_word)
    return ' '.join(formatted_words)

dim_table['Region'] = dim_table['Region'].apply(format_region)

dim_table
```

```python
dim_table['Location_ID'] = range(1, len(dim_table) + 1)

dim_table
```

```python
dim_table = dim_table[['Location_ID', 'Region']]

dim_table
```

```python
dim_table.to_csv('Tourism_Location_Dim.csv', index = False)
```
