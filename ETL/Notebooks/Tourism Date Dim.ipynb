```python
import pandas as pd
# import numpy as np

folder_path = 'DSS_DATASET_UPDATED.xlsx'
df = pd.read_excel(folder_path, sheet_name = 'O2')
# O2 = Purpose of Trips 
# O3 =  Number by Destination - Region
# O5 = Mode of Departure

df
```

```python
col_list = ['Unnamed: 1', 'Unnamed: 2']

dim_table = df[col_list]

dim_table
```

```python
dim_table.columns = ['Year', 'Month']
dim_table
```

```python
dim_table['Month']
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
index = 0
for month in dim_table['Month']:
    if month not in months:
        dim_table = dim_table.drop(index)
    index += 1
dim_table
```

```python
dim_table = dim_table.reset_index(drop=True)
dim_table
```

```python
dim_table.dtypes
```

```python
years = []
count = 0
for Num in dim_table['Year']:
    if not pd.isnull(Num):
        years.append(int(Num))
    else:
        if count > 0:
            years.append(int(years[count-1]))
        else:
            years.append(0)
    count += 1

dim_table['Year'] = years

dim_table
```

```python
quarters = []
for month in dim_table['Month']:
    if month in months[0:3]:
        quarters.append('Q1')
    elif month in months[3:6]:
        quarters.append('Q2')
    elif month in months[6:9]:
        quarters.append('Q3')
    elif month in months[9:]:
        quarters.append('Q4')
    else:
        quarters.append(None)
        # quarters.append(np.nan)

dim_table['Quarter'] = quarters

dim_table
```

```python
dim_table['Date_ID'] = range(1, len(dim_table) + 1)
dim_table
```

```python
dim_table = dim_table[['Date_ID', 'Year', 'Quarter', 'Month']]

dim_table
```

```python
dim_table.to_csv('Tourism_Date_Dim.csv', index = False)
```
