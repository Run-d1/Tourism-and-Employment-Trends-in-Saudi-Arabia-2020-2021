```python
import pandas as pd

folder_path = 'DSS_DATASET_UPDATED.xlsx'
df = pd.read_excel(folder_path, sheet_name = 'O5')
# O2 = Purpose of Trips 
# O3 =  Number by Destination - Region
# O5 = Mode of Departure

df
```

```python
df = df.transpose()

df
```

```python
col_list = [3]

dim_table = df[col_list]

dim_table
```

```python
dim_table.columns = ['Mode_of_Departure']

dim_table
```

```python
dim_table = dim_table.drop(dim_table.index[0:3]).reset_index(drop=True)

dim_table
```

```python
dim_table = dim_table.dropna().reset_index(drop=True)

dim_table
```

```python
dim_table = dim_table.drop(dim_table.index[3:])

dim_table
```

```python
dim_table['Mode_of_Departure_ID'] = range(1, len(dim_table) + 1)

dim_table
```

```python
dim_table = dim_table[['Mode_of_Departure_ID', 'Mode_of_Departure']]

dim_table
```

```python
dim_table['Mode_of_Departure'] = dim_table['Mode_of_Departure'].str.split('\n').str[0]

dim_table
```

```python
dim_table.to_csv('Tourism_Mode_of_Departure_Dim.csv', index=False)
```
