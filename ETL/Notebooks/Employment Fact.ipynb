```python
import pandas as pd

df = pd.read_csv('EmploymentDataset.csv')

df
```

```python
del df['Unnamed: 3']
df = df.drop(columns=['Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'])

df
```

```python
df = df.transpose().reset_index(drop=True)

df
```

```python
df[0] = df[0].ffill()

df
```

```python
df['Citizenship - Gender'] = df[0] + ' ' + df[1]

col = df['Citizenship - Gender']
df.drop(labels=['Citizenship - Gender'], axis=1, inplace=True)
df.insert(0, 'Citizenship - Gender', col)

df.drop([0, 1], axis=1, inplace=True)

df
```

```python
df = df.transpose().reset_index(drop=True)

df
```

```python
df.columns = df.iloc[0]
df = df[1:].reset_index(drop=True)

df
```

```python
df.columns = ['Year - Quarter', 'Saudi Male', 'Saudi Female', 'Non-Saudi Male', 'Non-Saudi Female']

df
```

```python
fact_table = pd.melt(df, id_vars=['Year - Quarter'], var_name="Citizenship and Gender", value_name="Number_of_Employment")

fact_table
```

```python
fact_table['Citizenship'] = fact_table['Citizenship and Gender'].str.split(' ').str[0]
fact_table['Gender'] = fact_table['Citizenship and Gender'].str.split(' ').str[1]

fact_table.drop(columns=['Citizenship and Gender'], inplace=True)

fact_table
```

```python
gender_dim = pd.read_csv('Emp_Gender_Dim.csv')

gender_dim
```

```python
fact_table = pd.merge(fact_table, gender_dim, on='Gender', how='left')

fact_table
```

```python
citizenship_dim = pd.read_csv('Tourism_Citizenship_Dim.csv')

citizenship_dim
```

```python
fact_table = pd.merge(fact_table, citizenship_dim, on='Citizenship', how='left')

fact_table
```

```python
date_dim = pd.read_csv('Tourism_Date_Dim.csv')

date_dim
```

```python
date_dim.drop("Month", axis=1, inplace=True)

date_dim
```

```python
date_dim.drop_duplicates(subset=["Year", "Quarter"], inplace=True)

date_dim
```

```python
date_dim.dtypes
```

```python
fact_table['Year'] = fact_table['Year - Quarter'].str.extract(r'(\d{4})')

fact_table
```

```python
fact_table['Quarter'] = fact_table['Year - Quarter'].str.extract(r'الربع (\S+)')

fact_table
```

```python
quarter_mapping = {
    'الأول': 'Q1',
    'الثاني': 'Q2',
    'الثالث': 'Q3',
    'الرابع': 'Q4'
}
fact_table.loc[:, 'Quarter'] = fact_table['Quarter'].map(quarter_mapping)

fact_table
```

```python
fact_table.dtypes
```

```python
fact_table['Year'] = fact_table['Year'].astype(int) 

fact_table
```

```python
fact_table = fact_table.merge(date_dim, on=['Year', 'Quarter'], how='left')

fact_table
```

```python
fact_table = fact_table[['Date_ID', 'Citizenship_ID', 'Gender_ID', 'Number_of_Employment']]

fact_table
```

```python
fact_table = fact_table.sort_values(by=['Date_ID', 'Citizenship_ID', 'Gender_ID'])

fact_table
```

```python
fact_table.to_csv('Employment_Fact.csv', index=False)
```
