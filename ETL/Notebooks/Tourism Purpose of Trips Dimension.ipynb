```python
import pandas as pd

folder_path = 'DSS_DATASET_UPDATED.xlsx'
df = pd.read_excel(folder_path, sheet_name = 'O2')
# O2 = Purpose of Trips 
# O3 =  Number by Destination - Region
# O5 = Mode of Departure

df
```

```python
df = df.transpose()
df
```

```python
col = [3]

dim_table = df[col]

dim_table.reset_index(drop=True)
```

```python
dim_table.columns = ['Purpose']
dim_table
```

```python
dim_table = dim_table.dropna().reset_index(drop=True)

dim_table
```

```python
dim_table = dim_table.drop(0)
dim_table = dim_table.drop(dim_table.index[4:])
dim_table
```

```python
dim_table['Purpose'] = dim_table['Purpose'].str.replace('*', '')

dim_table
```

```python
dim_table['Purpose_ID'] = range(1, len(dim_table) + 1)

dim_table
```

```python
dim_table = dim_table[['Purpose_ID', 'Purpose']]

dim_table
```

```python
dim_table['Purpose'] = dim_table['Purpose'].str.split('\n').str[0]

dim_table
```

```python
dim_table.to_csv('Tourism_Purpose_Dim.csv', index = False)
```
