```python
import pandas as pd
folder_path = 'EmploymentDataset.csv'
df = pd.read_csv(folder_path)
df
```

```python
df = df.transpose().reset_index(drop=True)

df
```

```python
col = [1]
gender_dim_df = df[col]
gender_dim_df
```

```python
gender_dim_df = gender_dim_df.rename(columns={1: 'Gender'})

gender_dim_df
```

```python
gender_mapping = {
    'الذكور': 'Male',
    'الاناث': 'Female'
}
gender_dim_df.loc[:, 'Gender'] = gender_dim_df['Gender'].map(gender_mapping)

gender_dim_df
```

```python
gender_dim_df = gender_dim_df.drop_duplicates().dropna().reset_index(drop=True)

gender_dim_df
```

```python
gender_dim_df['Gender_ID'] = range(1, len(gender_dim_df) + 1)
gender_dim_df
```

```python
gender_dim_df = gender_dim_df[['Gender_ID', 'Gender']]
gender_dim_df
```

```python
gender_dim_df.to_csv('Emp_Gender_Dim.csv', index = False)
```
